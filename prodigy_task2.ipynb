```python
def m_age(c):
    Age=c[0]
    Pclass=c[1]
    if pd.isnull(Age):
       if Pclass==1:
          return 38
       elif Pclass==2:
           return 29
       else:
           return 25
           
    else:
        return(Age)
```

```python
train_df["Age"]=train_df[["Age","Pclass"]].apply(m_age, axis=1)
sns.heatmap(train_df.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
train_df.drop("Cabin",axis=1, inplace=True)
sns.heatmap(train_df.isnull(),cbar=False, cmap="viridis")
```
